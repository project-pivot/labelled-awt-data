```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('AWT_data_Iris_March_2023_labelled_step5.csv',sep=";", encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
df
```

```python
# Convert 'Begin' column to datetime
df['Begin'] = pd.to_datetime(df['Begin'], format="%d-%m-%y %H:%M:%S")
df['End'] = pd.to_datetime(df['End'], format="%d-%m-%y %H:%M:%S")

# Create a new column 'Begin_Date' with just the date component
df['Begin_Date'] = df['Begin'].dt.date

# Display the DataFrame to verify the changes
df
```

```python
# Sort the DataFrame by Begin to ensure consecutive rows are adjacent
df = df.sort_values(by='Begin')

merged_rows = []
current_merge = None

for index, row in df.iterrows():
    if current_merge is None:
        # Start a new merge set
        current_merge = row
    elif (current_merge['End'] == row['Begin'] and
          current_merge['Activity'] == row['Activity'] and
          current_merge['Case'] == row['Case'] and
          current_merge['Begin_Date'] == row['Begin_Date']):
        # Merge current row into the merge set
        current_merge['End'] = row['End']
    else:
        # End current merge set, append to merged_rows
        merged_rows.append(current_merge)
        current_merge = row

# Append the last merge set
if current_merge is not None:
    merged_rows.append(current_merge)

# Create a new DataFrame from the merged rows
merged_df = pd.DataFrame(merged_rows)

# Drop unnecessary columns
merged_df = merged_df.drop(columns=['App', 'Title', 'Begin_Date'])

merged_df
```

```python
# Write the merged DataFrame to a new CSV file
merged_df.to_csv('AWT_data_Iris_March_2023_labelled_step6.csv', index=False)
```

```python
# Calculate duration and store it in a new column
merged_df['Duration'] = merged_df['End'] - merged_df['Begin']

merged_df
```

```python
# Write the dataframe to a new CSV file
merged_df.to_csv('AWT_data_Iris_March_2023_labelled_step7.csv', index=False)
```
